```python
import pickle
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

merged_relations_clean = [
    'antonym',
    'at location',
    'capable of',
    'causes',
    'created by',
    'is a',
    'desires',
    'has subevent',
    'part of',
    'has context',
    'has property',
    'made of',
    'not capable of',
    'not desires',
    'receives action',
    'related to',
    'used for',
]


# ── Helpers to load data ─────────────────────────────────────────────────────
def load_adj_data(path):
    with open(path, "rb") as f:
        return pickle.load(f)

def load_concept_vocab(path):
    with open(path, "r", encoding="utf-8") as f:
        return [line.strip() for line in f]

# ── 1) Extract the QA subgraph as a MultiDiGraph ───────────────────────────
def extract_subgraph(adj_entry, concept_vocab):
    adj       = adj_entry['adj']            # (R*N, N) sparse
    concepts  = adj_entry['concepts']       # length-N array of IDs
    qmask     = adj_entry['qmask']
    amask     = adj_entry['amask']

    R = adj.toarray().shape[0] // len(concepts)
    N = len(concepts)
    A = adj.toarray().reshape(R, N, N)

    G = nx.MultiDiGraph()
    # add nodes with flags
    for idx, cid in enumerate(concepts):
        G.add_node(concept_vocab[cid],
                   is_q=bool(qmask[idx]),
                   is_a=bool(amask[idx]))

    # add directed edges with relation index
    for r in range(R):
        M = A[r]
        for i in range(N):
            for j in range(N):
                if M[i, j]:
                    u = concept_vocab[concepts[i]]
                    v = concept_vocab[concepts[j]]
                    G.add_edge(u, v, relation=r)
    return G

# ── 2) Visualize with circular layout & size answers by connectivity ──────
def visualize_circular_layout(G, id2relation, title):
    plt.figure(figsize=(12, 12))

    # compute circular layout
    pos = nx.circular_layout(G)

    # categorize nodes
    question_nodes  = [n for n,d in G.nodes(data=True) if d['is_q']]
    answer_nodes    = [n for n,d in G.nodes(data=True) if d['is_a']]
    intermediate_nodes = [n for n,d in G.nodes(data=True)
                          if not d['is_q'] and not d['is_a']]

    # base sizes
    q_size = 600
    m_size = 300
    # for answers: size = base + scale * (in_degree+out_degree)
    answer_sizes = []
    for n in answer_nodes:
        deg = G.in_degree(n) + G.out_degree(n)
        answer_sizes.append(300 + deg * 200)

    # draw nodes by group
    nx.draw_networkx_nodes(G, pos,
        nodelist=question_nodes,
        node_color="green",
        node_shape="s",
        node_size=q_size,
        alpha=0.8,
        label="Question"
    )
    nx.draw_networkx_nodes(G, pos,
        nodelist=answer_nodes,
        node_color="purple",
        node_shape="o",
        node_size=answer_sizes,
        alpha=0.8,
        label="Answer (size ∝ connectivity)"
    )
    nx.draw_networkx_nodes(G, pos,
        nodelist=intermediate_nodes,
        node_color="lightblue",
        node_shape="d",
        node_size=m_size,
        alpha=0.8,
        label="Intermediate"
    )

    # draw edges
    nx.draw_networkx_edges(
        G, pos,
        arrowstyle='->',
        arrowsize=10,
        edge_color='gray',
        alpha=0.6
    )

    # draw labels
    nx.draw_networkx_labels(G, pos, font_size=8)

    # edge labels (combine multi‐edges)
    edge_labels_multi = {}
    for u, v, data in G.edges(data=True):
        rel_txt = id2relation[data['relation']].replace("_", " ")
        edge_labels_multi.setdefault((u, v), []).append(rel_txt)
    edge_labels = {k: "\n".join(v) for k, v in edge_labels_multi.items()}

    nx.draw_networkx_edge_labels(
        G, pos,
        edge_labels=edge_labels,
        font_size=6,
        label_pos=0.5,
        rotate=False
    )

    plt.title(title, fontsize=14)
    plt.axis('off')
    plt.legend(scatterpoints=1)
    plt.tight_layout()
    plt.show()

# ── Main: load and visualize ────────────────────────────────────────────────
if __name__ == "__main__":
    graph_path   = "/content/KGSweetSpot/data/csqa/graph/dev.graph.adj.pk"
    vocab_path   = "/content/KGSweetSpot/data/cpnet/concept.txt"

    adj_data      = load_adj_data(graph_path)
    concept_vocab = load_concept_vocab(vocab_path)
    id2relation   = merged_relations_clean

    # choose question index
    question_index = 0

    G = extract_subgraph(adj_data[question_index], concept_vocab)
    visualize_circular_layout(
        G,
        id2relation,
        title=f"Subgraph for Question #{question_index}"
    )
```

```python
# --- Configure input path (edit this if needed) ---
INPUT_JSON = "/content/KGSweetSpot/Inference_results/data/inference_Mistral-7B-Instruct-v0.3.concept_net_No_Knowledge.dev.csqa.json"  # <- put your JSON path here

import json, os, math
from pathlib import Path

# --- Helpers ---
def load_items(path):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)
    if not isinstance(data, list):
        raise ValueError("Input JSON must be a list of records.")
    return data

def ensure_ids(items):
    """Force stable 1-based 'id' for every item, overwriting if necessary (per your request)."""
    for i, it in enumerate(items, start=1):
        it["id"] = str(i)

def get_pred_conf(it):
    """Prefer 'pred_conf'; else derive from 'probs' (max), else from 'probs_' (first block max)."""
    if isinstance(it, dict) and "pred_conf" in it and isinstance(it["pred_conf"], (int, float)):
        return float(it["pred_conf"])
    if isinstance(it, dict) and "probs" in it and isinstance(it["probs"], list) and it["probs"]:
        return float(max(it["probs"]))
    if isinstance(it, dict) and "probs_" in it and isinstance(it["probs_"], list) and it["probs_"]:
        # probs_ is list of blocks; take the mean across blocks, then max position
        try:
            import numpy as np
            arr = np.array(it["probs_"], dtype=float)  # (B, N)
            mean_probs = arr.mean(axis=0)               # (N,)
            return float(mean_probs.max())
        except Exception:
            pass
    return None

def bin_items(items, hi_thr=0.85, mid_lo=0.65):
    hi, mid, lo = [], [], []
    skipped = []
    for it in items:
        p = get_pred_conf(it)
        if p is None:
            skipped.append(it)
            continue
        if p > hi_thr:
            hi.append(it)
        elif p >= mid_lo:
            mid.append(it)
        else:
            lo.append(it)
    return hi, mid, lo, skipped

# --- Do the work ---
items = load_items(INPUT_JSON)

# Apply sequential ids starting from 1 (per your last instruction)
ensure_ids(items)

# Split into bins
hi, mid, lo, skipped = bin_items(items, hi_thr=0.85, mid_lo=0.65)

# Prepare output paths
base = Path(INPUT_JSON)
stem = base.stem
out_dir = base.parent

paths = {
    "high": out_dir / f"{stem}.bin_high_gt0.85.json",
    "mid":  out_dir / f"{stem}.bin_mid_0.65_to_0.85.json",
    "low":  out_dir / f"{stem}.bin_low_lt0.65.json",
}

with open(paths["high"], "w", encoding="utf-8") as f:
    json.dump(hi, f, indent=2, ensure_ascii=False)
with open(paths["mid"], "w", encoding="utf-8") as f:
    json.dump(mid, f, indent=2, ensure_ascii=False)
with open(paths["low"], "w", encoding="utf-8") as f:
    json.dump(lo, f, indent=2, ensure_ascii=False)

summary = {
    "input_file": str(base),
    "total_items": len(items),
    "bins": {
        "high_gt_0.85": len(hi),
        "mid_0.65_to_0.85": len(mid),
        "low_lt_0.65": len(lo),
    }
}
summary
```

```python
# %%
from __future__ import annotations

import json
import math
import os
from dataclasses import dataclass
from typing import Dict, List, Tuple, Optional, Any

import numpy as np
import pandas as pd


# =========================
# Configuration (edit me!)
# =========================
DISPLAY_MODE = "std"     # (kept for compatibility; no longer affects Acceptance/Lift)
AS_PERCENT   = True      # show values as percentages
DECIMALS     = 2         # rounding for display
OUT_PATH     = "/content/Results_table.xlsx"  # change if needed

# Your mapping (exactly as you shared). Paths should point to your JSON dumps.
groups_by_model = {
    "Flan-T5-small": [
        {
            "name": "Conf < 0.65",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.concept_net_No_Knowledge.dev.csqa.bin_low_lt0.65.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.rival_rank5.json",
            )
        },
        {
            "name": "0.65 ≤ Conf ≤ 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.concept_net_No_Knowledge.dev.csqa.bin_mid_0.65_to_0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank5.json",
            )
        },
        {
            "name": "Conf > 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.concept_net_No_Knowledge.dev.csqa.bin_high_gt0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank5.json",
            )
        },
    ],
    "Flan-T5-large": [
        {
            "name": "Conf < 0.65",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.concept_net_No_Knowledge.dev.csqa.bin_low_lt0.65.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.rival_rank5.json",
            )
        },
        {
            "name": "0.65 ≤ Conf ≤ 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.concept_net_No_Knowledge.dev.csqa.bin_mid_0.65_to_0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank5.json",
            )
        },
        {
            "name": "Conf > 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.concept_net_No_Knowledge.dev.csqa.bin_high_gt0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank5.json",
            )
        },
    ],
    "Llama-2-7b-chat-hf": [
        {
            "name": "Conf < 0.65",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.concept_net_No_Knowledge.dev.csqa.bin_low_lt0.65.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank5.json",
            )
        },
        {
            "name": "0.65 ≤ Conf ≤ 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.concept_net_No_Knowledge.dev.csqa.bin_mid_0.65_to_0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank5.json",
            )
        },
        {
            "name": "Conf > 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.concept_net_No_Knowledge.dev.csqa.bin_high_gt0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank5.json",
            )
        },
    ],
    "Mistral-7B-Instruct-v0": [
        {
            "name": "Conf < 0.65",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.concept_net_No_Knowledge.dev.csqa.bin_low_lt0.65.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.rival_rank5.json",
            )
        },
        {
            "name": "0.65 ≤ Conf ≤ 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.concept_net_No_Knowledge.dev.csqa.bin_mid_0.65_to_0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank5.json",
            )
        },
        {
            "name": "Conf > 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.concept_net_No_Knowledge.dev.csqa.bin_high_gt0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank5.json",
            )
        },
    ]
}

# =========================
# Metrics & table builders
# =========================
@dataclass
class Metrics:
    acceptance: float
    lift: float
    adoption: float  # 0/1


def _index_by_choice(items: List[Dict[str, Any]]) -> Dict[str, float]:
    """Convert list of {'choice','prob'} to {choice: prob}."""
    return {d["choice"]: float(d["prob"]) for d in items}


# --- helpers to support missing `ranked` ---
def _has_ranked(item: Dict[str, Any]) -> bool:
    ranked = item.get("ranked")
    return isinstance(ranked, list) and all(("choice" in x and "prob" in x) for x in ranked)

def _ranked_from_base_probs(item: Dict[str, Any], sort_for_rank: bool) -> Optional[List[Dict[str, Any]]]:
    base_probs = item.get("base_probs")
    cands = item.get("cands")
    if not (isinstance(base_probs, list) and isinstance(cands, list) and len(base_probs) == len(cands)):
        return None
    pairs = [{"choice": cands[i], "prob": float(base_probs[i])} for i in range(len(cands))]
    if sort_for_rank:
        pairs = sorted(pairs, key=lambda d: d["prob"], reverse=True)
    return pairs


def _find_target_choice(item: Dict[str, Any], scenario_key: str) -> Optional[str]:
    """Target: base rank-1 for support; base rank k for r2..r5."""
    ranked_for_rank = item.get("ranked") if _has_ranked(item) else None
    if ranked_for_rank is None:
        ranked_for_rank = _ranked_from_base_probs(item, sort_for_rank=True)
    if not ranked_for_rank:
        return None

    if scenario_key == "support":
        base_pred = item.get("base_pred")
        return base_pred if isinstance(base_pred, str) else ranked_for_rank[0]["choice"]
    if scenario_key.startswith("r"):
        try:
            r = int(scenario_key[1:])
        except Exception:
            return None
        idx = r - 1
        return ranked_for_rank[idx]["choice"] if 0 <= idx < len(ranked_for_rank) else None
    return None


def _extract_acceptance_lift_adoption(item: Dict[str, Any], scenario_key: str) -> Optional[Metrics]:
    """Compute Acceptance, Lift, Adoption for a single record."""
    target_choice = _find_target_choice(item, scenario_key)
    if not target_choice:
        return None

    base_ranked = item.get("ranked") if _has_ranked(item) else _ranked_from_base_probs(item, sort_for_rank=False)
    if not base_ranked:
        return None

    base_dict = _index_by_choice(base_ranked)
    base_prob_target = float(base_dict.get(target_choice, np.nan))
    if math.isnan(base_prob_target):
        return None

    post_ranked = item.get("ranked_with_ks")
    if isinstance(post_ranked, list) and all("choice" in x and "prob" in x for x in post_ranked):
        post_dict = _index_by_choice(post_ranked)
    else:
        probs = item.get("probs")
        if isinstance(probs, list) and len(probs) == len(base_ranked):
            post_dict = {base_ranked[i]["choice"]: float(probs[i]) for i in range(len(base_ranked))}
        else:
            return None

    acceptance = float(post_dict.get(target_choice, np.nan))
    if math.isnan(acceptance):
        return None

    post_top = item.get("pred")
    if not isinstance(post_top, str):
        post_top = max(post_dict.items(), key=lambda kv: kv[1])[0]
    adoption = 1.0 if (post_top == target_choice) else 0.0

    lift = acceptance - base_prob_target
    return Metrics(acceptance=acceptance, lift=lift, adoption=adoption)


# ---------- base-confidence aggregation ----------
def _read_json(path: str) -> Optional[List[Dict[str, Any]]]:
    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
        return data if isinstance(data, list) else None
    except Exception:
        return None

def _extract_base_conf(item: Dict[str, Any]) -> Optional[float]:
    """Return the base confidence of the base prediction for this item."""
    x = item.get("base_pred_conf", None)
    if isinstance(x, (int, float)):
        return float(x)
    base_probs = item.get("base_probs")
    if isinstance(base_probs, list) and len(base_probs) > 0:
        try:
            return float(np.nanmax([float(v) for v in base_probs]))
        except Exception:
            pass
    ranked = item.get("ranked")
    if isinstance(ranked, list) and len(ranked) > 0 and all("prob" in r for r in ranked):
        try:
            return float(max(float(r["prob"]) for r in ranked))
        except Exception:
            pass
    return None

def _aggregate_base_confidence(base_path: str) -> Tuple[int, Dict[str, float]]:
    """Load base JSON and return (N_non_nan, {'median','q1','q3'})."""
    data = _read_json(base_path)
    if not data:
        return 0, {"median": float("nan"), "q1": float("nan"), "q3": float("nan")}
    vals: List[float] = []
    for item in data:
        v = _extract_base_conf(item)
        if isinstance(v, (int, float)) and not math.isnan(v):
            vals.append(float(v))
    if not vals:
        return 0, {"median": float("nan"), "q1": float("nan"), "q3": float("nan")}
    arr = np.array(vals, dtype=float)
    med = float(np.nanmedian(arr))
    q1  = float(np.nanpercentile(arr, 25))
    q3  = float(np.nanpercentile(arr, 75))
    return len(vals), {"median": med, "q1": q1, "q3": q3}


# ---------- stats & formatting ----------
def _median_iqr(arr: np.ndarray) -> Tuple[float, float, float]:
    if arr.size == 0:
        return (np.nan, np.nan, np.nan)
    med = float(np.nanmedian(arr))
    q1  = float(np.nanpercentile(arr, 25))
    q3  = float(np.nanpercentile(arr, 75))
    return med, q1, q3

def _mean_std(arr: np.ndarray) -> Tuple[float, float]:
    if arr.size == 0:
        return (np.nan, np.nan)
    mu = float(np.nanmean(arr))
    std = float(np.nanstd(arr, ddof=1)) if np.sum(~np.isnan(arr)) > 1 else 0.0
    return mu, std

def _summarize_metrics(metrics_list: List[Metrics]) -> Dict[str, Dict[str, float]]:
    if not metrics_list:
        return {
            "acceptance": {"median": np.nan, "q1": np.nan, "q3": np.nan},
            "lift":       {"median": np.nan, "q1": np.nan, "q3": np.nan},
            "adoption":   {"mean": np.nan,   "std": np.nan},
        }
    arr_acc = np.array([m.acceptance for m in metrics_list], dtype=float)
    arr_lft = np.array([m.lift for m in metrics_list], dtype=float)
    arr_adp = np.array([m.adoption for m in metrics_list], dtype=float)
    med_acc, q1_acc, q3_acc = _median_iqr(arr_acc)
    med_lft, q1_lft, q3_lft = _median_iqr(arr_lft)
    mean_adp, std_adp = _mean_std(arr_adp)
    return {
        "acceptance": {"median": med_acc, "q1": q1_acc, "q3": q3_acc},
        "lift":       {"median": med_lft, "q1": q1_lft, "q3": q3_lft},
        "adoption":   {"mean": mean_adp,  "std": std_adp},
    }

def _format_median_iqr(median: float, q1: float, q3: float, as_percent: bool, decimals: int) -> str:
    if any(math.isnan(x) for x in [median, q1, q3]):
        return "NA"
    scale = 100.0 if as_percent else 1.0
    m   = round(median * scale, decimals)
    q1s = round(q1 * scale,     decimals)
    q3s = round(q3 * scale,     decimals)
    return f"{m} [{q1s}–{q3s}]" + ("%" if as_percent else "")

def _format_mean_pm(mean: float, std: float, as_percent: bool, decimals: int) -> str:
    if any(math.isnan(x) for x in [mean, std]):
        return "NA"
    scale = 100.0 if as_percent else 1.0
    mu = round(mean * scale, decimals)
    sd = round(std  * scale, decimals)
    return f"{mu} ± {sd}" + ("%" if as_percent else "")


def _aggregate_condition(path: str, scenario_key: str) -> Tuple[int, Dict[str, Dict[str, float]]]:
    data = _read_json(path)
    if not data:
        return 0, {
            "acceptance": {"median": np.nan, "q1": np.nan, "q3": np.nan},
            "lift":       {"median": np.nan, "q1": np.nan, "q3": np.nan},
            "adoption":   {"mean": np.nan,   "std": np.nan},
        }
    metrics: List[Metrics] = []
    for item in data:
        m = _extract_acceptance_lift_adoption(item, scenario_key)
        if m is not None:
            metrics.append(m)
    stats = _summarize_metrics(metrics)
    return len(metrics), stats


def build_results_table(groups_by_model: Dict[str, List[Dict[str, Any]]],
                        display: str = "std",
                        as_percent: bool = False,
                        decimals: int = 2) -> pd.DataFrame:
    """
    Build MultiIndex DataFrame:
      rows: (Model, Confidence Bin, N)
      first column group: Base → Base Confidence (Median [Q1–Q3])
      then per scenario:
         - Acceptance (Median [Q1–Q3])
         - Lift       (Median [Q1–Q3])
         - Adoption   (Mean ± σ)
    """
    scenarios = [
        ("support", "Support@Rank1"),
        ("r2", "Rival@Rank2"),
        ("r3", "Rival@Rank3"),
        ("r4", "Rival@Rank4"),
        ("r5", "Rival@Rank5"),
    ]

    columns: List[Tuple[str, str]] = [("Base", "Base Confidence (Median [Q1–Q3])")]
    for _, scen_label in scenarios:
        columns.append((scen_label, "Acceptance (Median [Q1–Q3])"))
        columns.append((scen_label, "Lift (Median [Q1–Q3])"))
        columns.append((scen_label, "Adoption (Mean ± σ)"))
    columns = pd.MultiIndex.from_tuples(columns)

    rows = []
    values = []

    for model, bins in groups_by_model.items():
        for bin_info in bins:
            bin_name = bin_info["name"]
            paths = bin_info["paths"]

            row_values: List[str] = []
            Ns: List[int] = []

            # --- Base confidence median [Q1–Q3] ---
            base_path = paths.get("base")
            if base_path and os.path.exists(base_path):
                N_base, base_stats = _aggregate_base_confidence(base_path)
                base_str = _format_median_iqr(base_stats["median"], base_stats["q1"], base_stats["q3"], as_percent, decimals)
            else:
                N_base, base_str = 0, "NA"
            row_values.append(base_str)

            # --- Scenarios ---
            for scen_key, _ in scenarios:
                json_path = paths.get(scen_key)
                if not json_path or not os.path.exists(json_path):
                    Ns.append(0)
                    row_values.extend(["NA", "NA", "NA"])
                    continue

                N, stats = _aggregate_condition(json_path, scen_key)
                Ns.append(N)

                # Acceptance: median [Q1–Q3]
                a = stats["acceptance"]
                row_values.append(_format_median_iqr(a["median"], a["q1"], a["q3"], as_percent, decimals))

                # Lift: median [Q1–Q3]
                l = stats["lift"]
                row_values.append(_format_median_iqr(l["median"], l["q1"], l["q3"], as_percent, decimals))

                # Adoption: mean ± SD
                d = stats["adoption"]
                row_values.append(_format_mean_pm(d["mean"], d["std"], as_percent, decimals))

            # N = min across scenarios (unchanged)
            N_row = min(Ns) if Ns else 0
            rows.append((model, bin_name, N_row))
            values.append(row_values)

    index = pd.MultiIndex.from_tuples(rows, names=["Model", "Confidence Bin", "N"])
    df = pd.DataFrame(values, index=index, columns=columns)
    return df


def save_table(df: pd.DataFrame, out_path: str) -> None:
    """Save to Excel with frozen panes and approximate auto-fit widths."""
    from openpyxl.utils import get_column_letter

    out_dir = os.path.dirname(out_path)
    if out_dir and not os.path.exists(out_dir):
        os.makedirs(out_dir, exist_ok=True)

    with pd.ExcelWriter(out_path, engine="openpyxl") as writer:
        df.to_excel(writer, sheet_name="Results")
        ws = writer.book["Results"]

        # Freeze first row and first 3 columns (Model, Confidence Bin, N)
        ws.freeze_panes = "D2"

        # Approximate auto-fit: iterate by column index
        for col_idx in range(1, ws.max_column + 1):
            max_len = 0
            for row_idx in range(1, ws.max_row + 1):
                cell = ws.cell(row=row_idx, column=col_idx)
                val = "" if cell.value is None else str(cell.value)
                if len(val) > max_len:
                    max_len = len(val)
            col_letter = get_column_letter(col_idx)
            ws.column_dimensions[col_letter].width = min(max(12, max_len + 2), 40)


# =========================
# Build + Save
# =========================
df = build_results_table(groups_by_model, display=DISPLAY_MODE, as_percent=AS_PERCENT, decimals=DECIMALS)

# Show in notebook
try:
    from IPython.display import display
    display(df)
except Exception:
    print(df)

# Save to Excel
save_table(df, OUT_PATH)
print(f"\nSaved table to: {OUT_PATH}")
# Optionally also save CSV
csv_path = OUT_PATH.rsplit(".", 1)[0] + ".csv"
df.to_csv(csv_path)
print(f"Saved CSV to: {csv_path}")
```

```python
# Your mapping (exactly as you shared). Paths should point to your JSON dumps.
groups_by_model = {
    "Flan-T5-small": [
        {
            "name": "Conf < 0.65",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.concept_net_No_Knowledge.dev.csqa.bin_low_lt0.65.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_small/inference_flan-t5-small.rival_rank5.json",
            )
        },
        {
            "name": "0.65 ≤ Conf ≤ 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.concept_net_No_Knowledge.dev.csqa.bin_mid_0.65_to_0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank5.json",
            )
        },
        {
            "name": "Conf > 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.concept_net_No_Knowledge.dev.csqa.bin_high_gt0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_small/inference_flan-t5-small.rival_rank5.json",
            )
        },
    ],
    "Flan-T5-large": [
        {
            "name": "Conf < 0.65",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.concept_net_No_Knowledge.dev.csqa.bin_low_lt0.65.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Flan_T5_large/inference_flan-t5-large.rival_rank5.json",
            )
        },
        {
            "name": "0.65 ≤ Conf ≤ 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.concept_net_No_Knowledge.dev.csqa.bin_mid_0.65_to_0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank5.json",
            )
        },
        {
            "name": "Conf > 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.concept_net_No_Knowledge.dev.csqa.bin_high_gt0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Flan_T5_large/inference_flan-t5-large.rival_rank5.json",
            )
        },
    ],
    "Llama-2-7b-chat-hf": [
        {
            "name": "Conf < 0.65",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.concept_net_No_Knowledge.dev.csqa.bin_low_lt0.65.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank5.json",
            )
        },
        {
            "name": "0.65 ≤ Conf ≤ 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.concept_net_No_Knowledge.dev.csqa.bin_mid_0.65_to_0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank5.json",
            )
        },
        {
            "name": "Conf > 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.concept_net_No_Knowledge.dev.csqa.bin_high_gt0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Llama-2-7b-chat-hf/inference_Llama-2-7b-chat-hf.rival_rank5.json",
            )
        },
    ],
    "Mistral-7B-Instruct-v0": [
        {
            "name": "Conf < 0.65",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.concept_net_No_Knowledge.dev.csqa.bin_low_lt0.65.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_lower_0.65/Mistral-7B-Instruct-v0.3/inference_Mistral-7B-Instruct-v0.3.rival_rank5.json",
            )
        },
        {
            "name": "0.65 ≤ Conf ≤ 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.concept_net_No_Knowledge.dev.csqa.bin_mid_0.65_to_0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_between_0.65_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank5.json",
            )
        },
        {
            "name": "Conf > 0.85",
            "paths": dict(
                base="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.concept_net_No_Knowledge.dev.csqa.bin_high_gt0.85.json",
                support="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.knowledge_support_pred.json",
                r2="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank2.json",
                r3="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank3.json",
                r4="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank4.json",
                r5="/content/KGSweetSpot/Inference_results/data/Conf_higher_0.85/Mistral-7B-Instruct-v0/inference_Mistral-7B-Instruct-v0.3.rival_rank5.json",
            )
        },
    ]
}
# Flexible pastel palettes (including colorblind-friendly) and nicer styling.
# This cell defines multiple palettes and lets you pick one via THEME.
# Re-run to regenerate all figures with the selected palette.
#
# Files written:
#   /mnt/data/rank_trends_1x2.png
#   /mnt/data/resistance_acceptance_2x2_page1.png (and more pages as needed)
#   /mnt/data/resistance_adoption_2x2_page1.png  (and more pages as needed)

from __future__ import annotations

import json
import math
import os
from typing import Dict, List, Tuple, Optional, Any
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
from PIL import Image, ImageDraw

# ======================
# Palette configuration
# ======================
def _lighten(hex_color: str, factor: float = 0.25) -> str:
    """Lighten HEX color toward white by factor in [0,1]."""
    hex_color = hex_color.lstrip("#")
    r = int(hex_color[0:2], 16)
    g = int(hex_color[2:4], 16)
    b = int(hex_color[4:6], 16)
    r = int(r + (255 - r) * factor)
    g = int(g + (255 - g) * factor)
    b = int(b + (255 - b) * factor)
    return f"#{r:02x}{g:02x}{b:02x}"

# Base palettes
OKABE_ITO = ["#E69F00","#56B4E9","#009E73","#F0E442","#0072B2","#D55E00","#CC79A7","#000000"]
OKABE_ITO_SOFT = [_lighten(c, 0.35) for c in OKABE_ITO]  # softened, pastel-ish + colorblind-friendly

COOL_BLUES = ["#023e8a","#0077b6","#0096c7","#00b4d8","#48cae4","#90e0ef"]
COOL_BLUES_SOFT = [_lighten(c, 0.45) for c in COOL_BLUES]

MUTED_EARTH = ["#606c38","#283618","#bc6c25","#dda15e","#ccd5ae","#a3b18a"]
MUTED_EARTH_SOFT = [_lighten(c, 0.5) for c in MUTED_EARTH]

WARM_PASTELS = ["#cdb4db","#ffc8dd","#ffafcc","#bde0fe","#a2d2ff","#ffd6a5","#fdffb6","#cdeac0"]

PALETTES = {
    "okabe_ito_soft": OKABE_ITO_SOFT,
    "cool_blues_soft": COOL_BLUES_SOFT,
    "muted_earth_soft": MUTED_EARTH_SOFT,
    "warm_pastels": WARM_PASTELS,
}

# Pick your theme here:
THEME = "okabe_ito_soft"   # <- change to: "cool_blues_soft", "muted_earth_soft", or "warm_pastels"

# ----------------- Global Matplotlib aesthetics -----------------
rcParams.update({
    "figure.facecolor": "white",
    "axes.facecolor": "#fcfcfc",
    "axes.edgecolor": "#e3e3e3",
    "axes.linewidth": 0.8,
    "axes.titlesize": 13,
    "axes.labelsize": 12,
    "font.size": 11,
    "legend.fontsize": 9,
    "xtick.color": "#333333",
    "ytick.color": "#333333",
    "grid.color": "#e9e9e9",
    "grid.linewidth": 0.8,
    "savefig.dpi": 240,
})

# ----------------- Helpers to read / parse your JSONs -----------------
def _read_json(path: str) -> Optional[List[Dict[str, Any]]]:
    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
        return data if isinstance(data, list) else None
    except Exception:
        return None

def _has_ranked(item: Dict[str, Any]) -> bool:
    ranked = item.get("ranked")
    return isinstance(ranked, list) and all(("choice" in x and "prob" in x) for x in ranked)

def _ranked_from_base_probs(item: Dict[str, Any], sort_for_rank: bool) -> Optional[List[Dict[str, Any]]]:
    base_probs = item.get("base_probs")
    cands = item.get("cands")
    if not (isinstance(base_probs, list) and isinstance(cands, list) and len(base_probs) == len(cands)):
        return None
    pairs = [{"choice": cands[i], "prob": float(base_probs[i])} for i in range(len(cands))]
    if sort_for_rank:
        pairs = sorted(pairs, key=lambda d: d["prob"], reverse=True)
    return pairs

def _index_by_choice(items: List[Dict[str, Any]]) -> Dict[str, float]:
    return {d["choice"]: float(d["prob"]) for d in items}

def _find_target_choice(item: Dict[str, Any], scenario_key: str) -> Optional[str]:
    ranked_for_rank = item.get("ranked") if _has_ranked(item) else None
    if ranked_for_rank is None:
        ranked_for_rank = _ranked_from_base_probs(item, sort_for_rank=True)
    if not ranked_for_rank:
        return None

    if scenario_key == "support":
        base_pred = item.get("base_pred")
        return base_pred if isinstance(base_pred, str) else ranked_for_rank[0]["choice"]
    if scenario_key.startswith("r"):
        try:
            r = int(scenario_key[1:])
        except Exception:
            return None
        idx = r - 1
        return ranked_for_rank[idx]["choice"] if 0 <= idx < len(ranked_for_rank) else None
    return None

def _extract_acceptance_adoption(item: Dict[str, Any], scenario_key: str) -> Optional[Tuple[float, float]]:
    target_choice = _find_target_choice(item, scenario_key)
    if not target_choice:
        return None

    base_ranked = item.get("ranked") if _has_ranked(item) else _ranked_from_base_probs(item, sort_for_rank=False)
    if not base_ranked:
        return None

    post_ranked = item.get("ranked_with_ks")
    if isinstance(post_ranked, list) and all("choice" in x and "prob" in x for x in post_ranked):
        post_dict = _index_by_choice(post_ranked)
    else:
        probs = item.get("probs")
        if isinstance(probs, list) and len(probs) == len(base_ranked):
            post_dict = {base_ranked[i]["choice"]: float(probs[i]) for i in range(len(base_ranked))}
        else:
            return None

    acceptance = float(post_dict.get(target_choice, np.nan))
    if math.isnan(acceptance):
        return None

    post_top = item.get("pred")
    if not isinstance(post_top, str):
        post_top = max(post_dict.items(), key=lambda kv: kv[1])[0]
    adoption = 1.0 if (post_top == target_choice) else 0.0
    return acceptance, adoption

# ----------------- Summaries -----------------
SCEN_LABELS = {"support":"Support@Rank1","r2":"Rival@Rank2","r3":"Rival@Rank3","r4":"Rival@Rank4","r5":"Rival@Rank5"}
RIVAL_KEYS = ["r2", "r3", "r4", "r5"]
BIN_ORDER = ["Conf < 0.65", "0.65 ≤ Conf ≤ 0.85", "Conf > 0.85"]

def collect_summaries(groups_by_model: Dict[str, List[Dict[str, Any]]]) -> pd.DataFrame:
    summary_rows = []
    for model, bins in groups_by_model.items():
        for bin_info in bins:
            bin_name = bin_info["name"]
            paths = bin_info["paths"]
            for scen_key, scen_label in SCEN_LABELS.items():
                path = paths.get(scen_key)
                data = _read_json(path) if path and os.path.exists(path) else None
                acc_vals, adp_vals = [], []
                if data:
                    for item in data:
                        out = _extract_acceptance_adoption(item, scen_key)
                        if out is not None:
                            a, d = out
                            acc_vals.append(a)
                            adp_vals.append(d)
                acc_med = float(np.nanmedian(acc_vals)) if len(acc_vals) else np.nan
                adp_mean = float(np.nanmean(adp_vals)) if len(adp_vals) else np.nan
                summary_rows.append({
                    "model": model,
                    "bin": bin_name,
                    "scenario": scen_key,
                    "label": scen_label,
                    "acceptance_median": acc_med,
                    "adoption_mean": adp_mean,
                    "N": len(acc_vals)
                })
    df = pd.DataFrame(summary_rows)
    if not df.empty:
        df["bin"] = pd.Categorical(df["bin"], categories=BIN_ORDER, ordered=True)
        df = df.sort_values(["model", "bin", "scenario"])
    return df

# ----------------- Style helpers -----------------
def _ensure_dir(path: str):
    d = os.path.dirname(path)
    if d and not os.path.exists(d):
        os.makedirs(d, exist_ok=True)

def _beautify_axes(ax):
    ax.grid(True, axis="y")
    for s in ["top", "right"]:
        ax.spines[s].set_visible(False)
    for s in ["left", "bottom"]:
        ax.spines[s].set_color("#e3e3e3")

LINESTYLES = [
    (),           # solid
    (5, 2),       # dashed
    (3, 2, 1, 2), # dash-dot-ish
    (1, 1),       # dotted
]

def _line_kwargs(color: str) -> dict:
    return dict(
        color=color,
        linewidth=2.8,
        linestyle="-",           # <- solid
        marker="o",
        markersize=6.8,
        markerfacecolor="#ffffff",
        markeredgewidth=1.3,
        alpha=0.98,
    )

def _get_palette(theme: str, needed: int) -> List[str]:
    base = PALETTES.get(theme, PALETTES["okabe_ito_soft"])
    # Repeat if needed
    if needed <= len(base):
        return base[:needed]
    out = []
    k = len(base)
    for i in range(needed):
        out.append(base[i % k])
    return out

# ----------------- Single-chart renders (no subplots) -----------------
def chart_rival_vs_rank_all_models(df_sum: pd.DataFrame, metric: str, savepath: str, theme: str):
    assert metric in {"acceptance_median", "adoption_mean"}
    records = []
    models = sorted(df_sum["model"].dropna().unique())
    for model in models:
        for rk in [2,3,4,5]:
            scen = f"r{rk}"
            sub = df_sum[(df_sum["model"]==model) & (df_sum["scenario"]==scen)]
            vals = sub[metric].to_numpy(dtype=float)
            Ns   = sub["N"].to_numpy(dtype=float)
            if vals.size == 0 or np.nansum(Ns) == 0:
                agg = np.nan
            else:
                mask = (~np.isnan(vals)) & (~np.isnan(Ns)) & (Ns>0)
                agg = float(np.average(vals[mask], weights=Ns[mask])) if mask.any() else np.nan
            records.append({"model": model, "rival_rank": rk, metric: agg})
    dplot = pd.DataFrame(records)

    colors = _get_palette(theme, len(models))

    fig, ax = plt.subplots(figsize=(7.8, 5.0))
    for i, model in enumerate(models):
        d = dplot[dplot["model"]==model].sort_values("rival_rank")
        ax.plot(
            d["rival_rank"].to_numpy(),
            d[metric].to_numpy(),
            label=model,
            **_line_kwargs(colors[i])
        )
    ax.set_xticks([2,3,4,5])
    ax.set_xlabel("Rival Target Rank (2 = closest competitor)")
    if metric == "acceptance_median":
        ax.set_ylabel("Acceptance (median)")
        ax.set_title("Rival effect: Acceptance vs. rival rank (all models, N-weighted)")
    else:
        ax.set_ylabel("Adoption (mean)")
        ax.set_title("Rival effect: Adoption vs. rival rank (all models, N-weighted)")
    _beautify_axes(ax)
    ax.legend(title="Model", frameon=False, ncol=2)
    os.makedirs(os.path.dirname(savepath), exist_ok=True) if os.path.dirname(savepath) else None
    plt.tight_layout()
    plt.savefig(savepath, bbox_inches="tight")
    plt.close()


def chart_resistance_per_model(df_sum: pd.DataFrame, metric: str, model: str, savepath: str, theme: str):
    assert metric in {"acceptance_median", "adoption_mean"}
    sub = df_sum[(df_sum["model"]==model) & (df_sum["scenario"].isin(["r2","r3","r4","r5"]))].copy()
    sub["bin"] = pd.Categorical(sub["bin"], categories=BIN_ORDER, ordered=True)
    _x = np.arange(len(BIN_ORDER))

    colors = _get_palette(theme, 4)  # one per rival rank
    fig, ax = plt.subplots(figsize=(6.6, 4.6))
    for i, scen in enumerate(["r2","r3","r4","r5"]):
        d = sub[sub["scenario"]==scen].sort_values("bin")
        y = d[metric].to_numpy(dtype=float)
        label = {"r2":"Rival@Rank2", "r3":"Rival@Rank3", "r4":"Rival@Rank4", "r5":"Rival@Rank5"}[scen]
        ax.plot(_x, y, label=label, **_line_kwargs(colors[i]))
    ax.set_xticks(_x)
    ax.set_xticklabels(["Low", "Moderate", "High"])
    ax.set_xlabel("Base confidence bin")
    if metric == "acceptance_median":
        ax.set_ylabel("Acceptance (median)")
        ax.set_title(f"Acceptance • {model}")
    else:
        ax.set_ylabel("Adoption (mean)")
        ax.set_title(f"Adoption • {model}")
    _beautify_axes(ax)
    ax.legend(title="Knowledge target", frameon=False)
    os.makedirs(os.path.dirname(savepath), exist_ok=True) if os.path.dirname(savepath) else None
    plt.tight_layout()
    plt.savefig(savepath, bbox_inches="tight")
    plt.close()
# ----------------- Collage builders (PIL) -----------------
def _open_or_blank(path: str, size=(980, 640)) -> Image.Image:
    if os.path.exists(path):
        im = Image.open(path).convert("RGB")
        im = im.resize(size, Image.LANCZOS)
    else:
        im = Image.new("RGB", size, "white")
    return im

def make_collage_1x2(left_path: str, right_path: str, out_path: str, pad: int=18, tile_size=(980, 640), title: Optional[str]=None):
    left  = _open_or_blank(left_path,  size=tile_size)
    right = _open_or_blank(right_path, size=tile_size)
    w = tile_size[0]*2 + pad*3
    h = tile_size[1] + pad*2 + (48 if title else 0)
    canvas = Image.new("RGB", (w, h), "white")
    y0 = pad + (48 if title else 0)
    canvas.paste(left,  (pad, y0))
    canvas.paste(right, (pad*2 + tile_size[0], y0))
    if title:
        draw = ImageDraw.Draw(canvas)
        draw.text((pad, pad), title, fill=(0,0,0))
    os.makedirs(os.path.dirname(out_path), exist_ok=True) if os.path.dirname(out_path) else None
    canvas.save(out_path)

def make_collage_2x2(img_paths: List[str], out_path: str, pad: int=18, tile_size=(980, 640), title: Optional[str]=None):
    tiles = []
    for i in range(4):
        p = img_paths[i] if i < len(img_paths) else None
        tiles.append(_open_or_blank(p or "", size=tile_size))
    w = tile_size[0]*2 + pad*3
    h = tile_size[1]*2 + pad*3 + (48 if title else 0)
    canvas = Image.new("RGB", (w, h), "white")
    y0 = pad + (48 if title else 0)
    canvas.paste(tiles[0], (pad, y0))
    canvas.paste(tiles[1], (pad*2 + tile_size[0], y0))
    canvas.paste(tiles[2], (pad, y0 + tile_size[1] + pad))
    canvas.paste(tiles[3], (pad*2 + tile_size[0], y0 + tile_size[1] + pad))
    if title:
        draw = ImageDraw.Draw(canvas)
        draw.text((pad, pad), title, fill=(0,0,0))
    os.makedirs(os.path.dirname(out_path), exist_ok=True) if os.path.dirname(out_path) else None
    canvas.save(out_path)

# ----------------- Driver -----------------
def run_all(groups_by_model: Dict[str, List[Dict[str, Any]]], theme: str = THEME):
    df_sum = collect_summaries(groups_by_model)
    if df_sum.empty:
        print("No data summarized. Check that your JSON paths exist and are readable.")
        return

    # 1) Rival effect rank trends (1x2 composite)
    p_acc = "/mnt/data/tmp_rank_acceptance.png"
    p_adp = "/mnt/data/tmp_rank_adoption.png"
    chart_rival_vs_rank_all_models(df_sum, metric="acceptance_median", savepath=p_acc, theme=theme)
    chart_rival_vs_rank_all_models(df_sum, metric="adoption_mean",    savepath=p_adp, theme=theme)
    out_rank = "/content/rank_trends_1x2.png"
    make_collage_1x2(p_acc, p_adp, out_rank, title="Rival effect trends (all models, N-weighted)")
    print(f"Saved 1x2 rank-trend figure: {out_rank}")

    # 2) Resistance grids per model (2x2 composites). Chunk models by 4.
    models = sorted(df_sum["model"].dropna().unique())
    chunks = [models[i:i+4] for i in range(0, len(models), 4)]

    pages_acc, pages_adp = [], []
    for page_idx, chunk in enumerate(chunks, start=1):
        tmp_paths_acc, tmp_paths_adp = [], []
        for m in chunk:
            pa = f"/mnt/data/tmp_resist_acc_{page_idx}_{m}.png"
            pd = f"/mnt/data/tmp_resist_adp_{page_idx}_{m}.png"
            chart_resistance_per_model(df_sum, metric="acceptance_median", model=m, savepath=pa, theme=theme)
            chart_resistance_per_model(df_sum, metric="adoption_mean",    model=m, savepath=pd, theme=theme)
            tmp_paths_acc.append(pa)
            tmp_paths_adp.append(pd)
        out_a = f"/content/resistance_acceptance_2x2_page{page_idx}.png"
        out_d = f"/content/resistance_adoption_2x2_page{page_idx}.png"
        make_collage_2x2(tmp_paths_acc, out_a, title="Resistance • Acceptance (Rival@2–5 vs base confidence)")
        make_collage_2x2(tmp_paths_adp, out_d, title="Resistance • Adoption (Rival@2–5 vs base confidence)")
        pages_acc.append(out_a); pages_adp.append(out_d)
        print(f"Saved 2x2 resistance figures (page {page_idx}):")
        print(" -", out_a)
        print(" -", out_d)

    return {
        "rank_trends": out_rank,
        "resistance_acceptance_pages": pages_acc,
        "resistance_adoption_pages": pages_adp,
    }

# Try to run (requires groups_by_model to be defined)
if 'groups_by_model' in globals():
    outputs = run_all(groups_by_model, theme=THEME)
    if outputs:
        print("\nArtifacts:")
        print(" Rank trends (1x2):", outputs["rank_trends"])
        for p in outputs["resistance_acceptance_pages"]:
            print(" Resistance acceptance:", p)
        for p in outputs["resistance_adoption_pages"]:
            print(" Resistance adoption:", p)
else:
    print("Ready. Set THEME and run: outputs = run_all(groups_by_model, theme='cool_blues_soft')")
```

```python
# %% Noise table with ONE base column and NO deltas
from __future__ import annotations

import json, os, math
from typing import Dict, List, Tuple, Optional, Any
import numpy as np
import pandas as pd

# ---------------- Config ----------------
AS_PERCENT = True
DECIMALS   = 2
OUT_PATH   = "/content/Noise_results_table.xlsx"

BASE_DIR = "/content/KGSweetSpot/Inference_results/data/noise_global"
NOISE_LEVELS = [10, 20, 30]   # scenarios to include
BIN_KEYS = ["low", "medium", "high"]
BIN_LABELS = {"low": "Conf < 0.65", "medium": "0.65 ≤ Conf ≤ 0.85", "high": "Conf > 0.85"}

MODEL_MAP = {
    "Flan-T5-small":          ("T5_small",        "inference_flan-t5-small"),
    "Flan-T5-large":          ("T5_large",        "inference_flan-t5-large"),
    "Llama-2-7b-chat-hf":     ("Llama-2-7b",      "inference_Llama-2-7b-chat-hf"),
    "Mistral-7B-Instruct-v0": ("Mistral-7B",      "inference_Mistral-7B-Instruct-v0.3"),
}

# If True → require all levels in NOISE_LEVELS to be present.
# If False → intersect across only the levels that actually exist.
REQUIRE_ALL_LEVELS = False

# --------------- Helpers ----------------
def _read_json(path: str) -> Optional[List[Dict[str, Any]]]:
    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
        return data if isinstance(data, list) else None
    except Exception:
        return None

def _has_ranked(item: Dict[str, Any]) -> bool:
    r = item.get("ranked")
    return isinstance(r, list) and all(("choice" in x and "prob" in x) for x in r)

def _ranked_from_base_probs(item: Dict[str, Any], sort_for_rank: bool) -> Optional[List[Dict[str, Any]]]:
    base_probs = item.get("base_probs"); cands = item.get("cands")
    if not (isinstance(base_probs, list) and isinstance(cands, list) and len(base_probs) == len(cands)):
        return None
    pairs = [{"choice": cands[i], "prob": float(base_probs[i])} for i in range(len(cands))]
    if sort_for_rank:
        pairs = sorted(pairs, key=lambda d: d["prob"], reverse=True)
    return pairs

def _index_by_choice(items: List[Dict[str, Any]]) -> Dict[str, float]:
    return {d["choice"]: float(d["prob"]) for d in items}

def _median_iqr(arr: np.ndarray) -> Tuple[float, float, float]:
    if arr.size == 0:
        return (np.nan, np.nan, np.nan)
    med = float(np.nanmedian(arr))
    q1  = float(np.nanpercentile(arr, 25))
    q3  = float(np.nanpercentile(arr, 75))
    return med, q1, q3

def _fmt_iqr(med: float, q1: float, q3: float, pct: bool, decimals: int) -> str:
    if any(math.isnan(x) for x in (med, q1, q3)): return "NA"
    s = 100.0 if pct else 1.0
    return f"{round(med*s,decimals)} [{round(q1*s,decimals)}–{round(q3*s,decimals)}]" + ("%" if pct else "")

# Parse one item → (base_choice, base_conf, post_conf_for_same_choice)
def _extract_base_and_post(item: Dict[str, Any]) -> Optional[Tuple[str, float, float]]:
    ranked_sorted = item.get("ranked") if _has_ranked(item) else _ranked_from_base_probs(item, True)
    ranked_for_map = item.get("ranked") if _has_ranked(item) else _ranked_from_base_probs(item, False)
    if not ranked_sorted or not ranked_for_map:
        return None

    base_choice = item.get("base_pred") if isinstance(item.get("base_pred"), str) else ranked_sorted[0]["choice"]
    base_dict = _index_by_choice(ranked_for_map)
    base_conf = float(base_dict.get(base_choice, np.nan))
    if math.isnan(base_conf): return None

    post_ranked = item.get("ranked_with_ks")
    if isinstance(post_ranked, list) and all(("choice" in x and "prob" in x) for x in post_ranked):
        post_dict = _index_by_choice(post_ranked)
    else:
        probs = item.get("probs")
        if isinstance(probs, list) and len(probs) == len(ranked_for_map):
            post_dict = {ranked_for_map[i]["choice"]: float(probs[i]) for i in range(len(ranked_for_map))}
        else:
            return None

    post_conf = float(post_dict.get(base_choice, np.nan))
    if math.isnan(post_conf): return None
    return base_choice, base_conf, post_conf

# Load a noise file → dict[id] = (base_conf, post_conf)
def _load_noise_file(path: str) -> Dict[str, Tuple[float, float]]:
    data = _read_json(path) or []
    out: Dict[str, Tuple[float, float]] = {}
    for item in data:
        rid = str(item.get("id"))
        parsed = _extract_base_and_post(item)
        if not rid or parsed is None:
            continue
        _, b, p = parsed
        out[rid] = (b, p)
    return out

# ------------- Build table --------------
def build_noise_table_one_base_no_delta() -> pd.DataFrame:
    # Columns: one Base, then per level Post (no Δ)
    cols = [("—", "Base Conf. (Median [Q1–Q3])")]
    for L in NOISE_LEVELS:
        lab = f"Noise@{L}"
        cols.append((lab, "Post-Noise Conf. (Median [Q1–Q3])"))
    columns = pd.MultiIndex.from_tuples(cols)

    rows, values = [], []
    for model_name, (subdir, stem) in MODEL_MAP.items():
        for binkey in BIN_KEYS:
            # Load all available levels
            level_dicts: Dict[int, Dict[str, Tuple[float, float]]] = {}
            for L in NOISE_LEVELS:
                path = os.path.join(BASE_DIR, f"{L}_noisy_knowledge", subdir, f"{stem}.noise_global.{binkey}.json")
                if os.path.exists(path):
                    d = _load_noise_file(path)
                    if d:
                        level_dicts[L] = d

            # Determine ID intersection
            if REQUIRE_ALL_LEVELS:
                present = [L for L in NOISE_LEVELS if L in level_dicts]
                ids_intersection = set.intersection(*(set(level_dicts[L].keys()) for L in NOISE_LEVELS)) if len(present)==len(NOISE_LEVELS) else set()
            else:
                present = [L for L in NOISE_LEVELS if L in level_dicts]
                ids_intersection = set.intersection(*(set(level_dicts[L].keys()) for L in present)) if present else set()

            row_cells: List[str] = []
            N = len(ids_intersection)

            if N > 0:
                # Base from the first present level
                firstL = present[0]
                base_vals = np.array([level_dicts[firstL][rid][0] for rid in ids_intersection], dtype=float)
                b_med, b_q1, b_q3 = _median_iqr(base_vals)
                row_cells.append(_fmt_iqr(b_med, b_q1, b_q3, AS_PERCENT, DECIMALS))
            else:
                row_cells.append("NA")

            # Post only (no deltas)
            for L in NOISE_LEVELS:
                if N == 0 or L not in level_dicts:
                    row_cells.append("NA")
                    continue
                post_vals = np.array([level_dicts[L][rid][1] for rid in ids_intersection], dtype=float)
                p_med, p_q1, p_q3 = _median_iqr(post_vals)
                row_cells.append(_fmt_iqr(p_med, p_q1, p_q3, AS_PERCENT, DECIMALS))

            rows.append((model_name, BIN_LABELS[binkey], N))
            values.append(row_cells)

    index = pd.MultiIndex.from_tuples(rows, names=["Model", "Confidence Bin", "N"])
    df = pd.DataFrame(values, index=index, columns=columns)
    return df

def save_table(df: pd.DataFrame, out_path: str) -> None:
    from openpyxl.utils import get_column_letter
    out_dir = os.path.dirname(out_path)
    if out_dir and not os.path.exists(out_dir): os.makedirs(out_dir, exist_ok=True)
    with pd.ExcelWriter(out_path, engine="openpyxl") as w:
        df.to_excel(w, sheet_name="Noise")
        ws = w.book["Noise"]
        ws.freeze_panes = "D2"
        for c in range(1, ws.max_column + 1):
            max_len = 0
            for r in range(1, ws.max_row + 1):
                v = ws.cell(row=r, column=c).value
                s = "" if v is None else str(v)
                if len(s) > max_len: max_len = len(s)
            ws.column_dimensions[get_column_letter(c)].width = min(max(12, max_len + 2), 42)

# -------- Run --------
df_noise = build_noise_table_one_base_no_delta()
try:
    from IPython.display import display
    display(df_noise)
except Exception:
    print(df_noise)

save_table(df_noise, OUT_PATH)
print(f"Saved noise table to: {OUT_PATH}")
csv_path = OUT_PATH.rsplit(".", 1)[0] + ".csv"
df_noise.to_csv(csv_path)
print(f"Saved CSV to: {csv_path}")
```

```python
# %%
from __future__ import annotations
import json, os
from typing import Dict, List, Any, Optional
import numpy as np
import matplotlib.pyplot as plt

# -------------------
# Config
# -------------------
ROOT = "/content/KGSweetSpot/Inference_results/data/noise_global"  # change if needed
K_LIST = [10, 20, 30]   # figures generated for each K
BINS   = ["low", "medium", "high"]  # columns
MODELS = [
    "Flan-T5-small",
    "Flan-T5-large",
    "Llama-2-7b-chat-hf",
    "Mistral-7B-Instruct-v0.3",
]
OUT_DIR = "/content/noise_figs"
os.makedirs(OUT_DIR, exist_ok=True)

# Colors (same as your T5 figs)
COL_BASE = "#7DB8FF"    # base (blue)
COL_NOIS = "#FF9E57"    # with-noise (orange)

# -------------------
# Paths
# -------------------
def _model_folder_and_prefix(model: str):
    if model == "Flan-T5-small":
        return "T5_small", "inference_flan-t5-small.noise_global"
    if model == "Flan-T5-large":
        return "T5_large", "inference_flan-t5-large.noise_global"
    if model == "Llama-2-7b-chat-hf":
        return "Llama-2-7b", "inference_Llama-2-7b-chat-hf.noise_global"
    if model == "Mistral-7B-Instruct-v0.3":
        return "Mistral-7B", "inference_Mistral-7B-Instruct-v0.3.noise_global"
    raise ValueError(model)

def build_path(K: int, model: str, bin_key: str) -> str:
    subfolder, prefix = _model_folder_and_prefix(model)
    return os.path.join(ROOT, f"{K}_noisy_knowledge", subfolder, f"{prefix}.{bin_key}.json")

# -------------------
# IO + helpers
# -------------------
def read_json(path: str) -> Optional[List[Dict[str, Any]]]:
    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
        return data if isinstance(data, list) else None
    except Exception as e:
        print(f"[WARN] read fail {path}: {e}")
        return None

def _has_ranked(item: Dict[str, Any]) -> bool:
    r = item.get("ranked")
    return isinstance(r, list) and all(("choice" in x and "prob" in x) for x in r)

def _ranked_from_base(item: Dict[str, Any], sort_for_rank: bool) -> Optional[List[Dict[str, Any]]]:
    base_probs = item.get("base_probs")
    cands = item.get("cands")
    if not (isinstance(base_probs, list) and isinstance(cands, list) and len(base_probs) == len(cands)):
        return None
    pairs = [{"choice": cands[i], "prob": float(base_probs[i])} for i in range(len(cands))]
    if sort_for_rank:
        pairs = sorted(pairs, key=lambda d: d["prob"], reverse=True)
    return pairs

def _index(items: List[Dict[str, Any]]) -> Dict[str, float]:
    return {d["choice"]: float(d["prob"]) for d in items}

# -------------------
# Per-rank medians
# -------------------
def per_rank_medians_for_file(path: str):
    """
    Returns two arrays of length 5:
      base_med[r-1], post_med[r-1]  for ranks r=1..5,
    where 'post' are probabilities of the SAME base-ranked choices after noise.
    """
    data = read_json(path)
    if not data:
        return None, None

    base_lists = {r: [] for r in range(1, 6)}
    post_lists = {r: [] for r in range(1, 6)}

    for item in data:
        # 1) base ranking (sorted by prob desc) to define rank positions
        ranked_sorted = item.get("ranked") if _has_ranked(item) else None
        if ranked_sorted is None:
            ranked_sorted = _ranked_from_base(item, sort_for_rank=True)
        if not ranked_sorted:
            continue

        # 2) post dictionary (prefer ranked_with_ks; otherwise probs aligned to base order)
        if isinstance(item.get("ranked_with_ks"), list):
            post_dict = _index(item["ranked_with_ks"])
        else:
            base_unordered = item.get("ranked") if _has_ranked(item) else _ranked_from_base(item, sort_for_rank=False)
            probs = item.get("probs")
            if not (base_unordered and isinstance(probs, list) and len(probs) == len(base_unordered)):
                continue
            post_dict = {base_unordered[i]["choice"]: float(probs[i]) for i in range(len(base_unordered))}

        # 3) collect by base rank
        for r in range(1, min(6, len(ranked_sorted)+1)):
            ch   = ranked_sorted[r-1]["choice"]
            bval = float(ranked_sorted[r-1]["prob"])
            pval = float(post_dict.get(ch, np.nan))
            if not np.isnan(bval) and not np.isnan(pval):
                base_lists[r].append(bval)
                post_lists[r].append(pval)

    base_med = [float(np.nanmedian(base_lists[r])) if len(base_lists[r]) else np.nan for r in range(1, 6)]
    post_med = [float(np.nanmedian(post_lists[r])) if len(post_lists[r]) else np.nan for r in range(1, 6)]
    return base_med, post_med

# -------------------
# Plot: 4 rows (models) × 3 cols (bins)
# -------------------
def plot_all_models_for_k(K: int, save_path: str) -> str:
    import math
    fig, axes = plt.subplots(len(MODELS), len(BINS), figsize=(15, 10), sharex=True, sharey=True)
    if len(MODELS) == 1:
        axes = np.array([axes])
    x = np.arange(1, 6)

    for i, model in enumerate(MODELS):
        for j, bin_key in enumerate(BINS):
            ax = axes[i, j]
            path = build_path(K, model, bin_key)
            base_med, post_med = per_rank_medians_for_file(path)

            if base_med is not None:
                ax.plot(x, base_med, marker="o", linewidth=2.0, color=COL_BASE, label="base")
            if post_med is not None:
                ax.plot(x, post_med, marker="o", linewidth=2.0, color=COL_NOIS, label="with-noise")

            if i == 0:
                ax.set_title(f"{bin_key.capitalize()}", fontsize=11)
            if j == 0:
                ax.set_ylabel(model, fontsize=10)
            ax.set_xticks(x)
            ax.set_xlabel("Base rank (1 = top)", fontsize=9)
            ax.grid(axis="y", linestyle=":", alpha=0.35)
            ax.set_ylim(0, 1.02)

    # single legend
    handles = [
        plt.Line2D([0],[0], color=COL_BASE, linewidth=2.5, marker="o"),
        plt.Line2D([0],[0], color=COL_NOIS, linewidth=2.5, marker="o"),
    ]
    labels = ["base", "with-noise"]
    fig.legend(handles, labels, loc="upper center", ncol=2, frameon=False, bbox_to_anchor=(0.5, 1.02))
    fig.suptitle(f"Median P vs base rank — Global noise (all models) — K={K}", y=1.06, fontsize=13)

    fig.tight_layout()
    fig.savefig(save_path, dpi=220, bbox_inches="tight")
    plt.close(fig)
    return save_path

# -------------------
# Run for each K
# -------------------
outs = []
for K in K_LIST:
    out = os.path.join(OUT_DIR, f"global_noise_rank_trends_all_models_K{K}.png")
    print("Saved:", plot_all_models_for_k(K, out))
    outs.append(out)

outs
# %%
```

```python
# %%
from __future__ import annotations

import json
import math
import os
from typing import Dict, List, Tuple, Optional, Any

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# Configuration
# =========================
ROOT = "/content/KGSweetSpot/Inference_results/data/noise_global"  # change if needed
KS = [10, 20, 30]  # number of random noisy statements
MODELS = [
    "Flan-T5-small",
    "Flan-T5-large",
    "Llama-2-7b-chat-hf",
    "Mistral-7B-Instruct-v0.3",
]
BINS = ["low", "medium", "high"]  # low:<0.65, medium:0.65-0.85, high:>0.85

OUT_DIR = "./noise_figs"  # where to save the summary CSV + figs
os.makedirs(OUT_DIR, exist_ok=True)

# Toggle: group by bin or aggregate all bins together
GROUP_BY_BIN = False  # set True to also plot per-bin charts

# =========================
# Path patterns
# =========================
def _model_folder_and_prefix(model: str) -> Tuple[str, str]:
    """
    Return (subfolder, filename_prefix) for a given model, as in your examples.
    """
    if model == "Flan-T5-small":
        return "T5_small", "inference_flan-t5-small.noise_global"
    if model == "Flan-T5-large":
        return "T5_large", "inference_flan-t5-large.noise_global"
    if model == "Llama-2-7b-chat-hf":
        return "Llama-2-7b", "inference_Llama-2-7b-chat-hf.noise_global"
    if model == "Mistral-7B-Instruct-v0.3":
        return "Mistral-7B", "inference_Mistral-7B-Instruct-v0.3.noise_global"
    raise ValueError(f"Unknown model: {model}")

def build_noise_path(K: int, model: str, bin_key: str) -> str:
    subfolder, prefix = _model_folder_and_prefix(model)
    return os.path.join(
        ROOT,
        f"{K}_noisy_knowledge",
        subfolder,
        f"{prefix}.{bin_key}.json",
    )

# =========================
# JSON helpers
# =========================
def read_json(path: str) -> Optional[List[Dict[str, Any]]]:
    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
        return data if isinstance(data, list) else None
    except Exception as e:
        print(f"[WARN] Failed to read {path}: {e}")
        return None

def _has_ranked(item: Dict[str, Any]) -> bool:
    ranked = item.get("ranked")
    return isinstance(ranked, list) and all(("choice" in x and "prob" in x) for x in ranked)

def _ranked_from_base_probs(item: Dict[str, Any], sort_for_rank: bool) -> Optional[List[Dict[str, Any]]]:
    base_probs = item.get("base_probs")
    cands = item.get("cands")
    if not (isinstance(base_probs, list) and isinstance(cands, list) and len(base_probs) == len(cands)):
        return None
    pairs = [{"choice": cands[i], "prob": float(base_probs[i])} for i in range(len(cands))]
    if sort_for_rank:
        pairs = sorted(pairs, key=lambda d: d["prob"], reverse=True)
    return pairs

def _index_by_choice(items: List[Dict[str, Any]]) -> Dict[str, float]:
    return {d["choice"]: float(d["prob"]) for d in items}

# =========================
# Core extraction for noisy
# =========================
def extract_base_and_post_conf(item: Dict[str, Any]) -> Optional[Tuple[float, float]]:
    """
    For a noisy item:
      - Find the base top-1 choice (b) from base (ranked or base_probs).
      - Base confidence = probability of b in base distribution.
      - Post confidence = probability of the same b in ranked_with_ks (preferred),
        else fallback to 'probs' aligned to base order.
    Returns (base_conf, post_conf) or None if not available.
    """
    # Find base ranking for picking top-1
    ranked_for_rank = item.get("ranked") if _has_ranked(item) else None
    if ranked_for_rank is None:
        ranked_for_rank = _ranked_from_base_probs(item, sort_for_rank=True)
    if not ranked_for_rank:
        return None

    base_top_choice = item.get("base_pred")
    if not isinstance(base_top_choice, str):
        base_top_choice = ranked_for_rank[0]["choice"]

    # Base distribution (for getting base probabilities by choice)
    base_ranked = item.get("ranked") if _has_ranked(item) else _ranked_from_base_probs(item, sort_for_rank=False)
    if not base_ranked:
        return None
    base_dict = _index_by_choice(base_ranked)
    base_conf = base_dict.get(base_top_choice, np.nan)
    if base_conf is None or math.isnan(base_conf):
        return None

    # Post distribution
    post_ranked = item.get("ranked_with_ks")
    if isinstance(post_ranked, list) and all(("choice" in x and "prob" in x) for x in post_ranked):
        post_dict = _index_by_choice(post_ranked)
        post_conf = post_dict.get(base_top_choice, np.nan)
    else:
        # fallback: use 'probs' aligned to the order of base_ranked
        probs = item.get("probs")
        if isinstance(probs, list) and len(probs) == len(base_ranked):
            # align by base order
            post_dict = {base_ranked[i]["choice"]: float(probs[i]) for i in range(len(base_ranked))}
            post_conf = post_dict.get(base_top_choice, np.nan)
        else:
            return None

    if post_conf is None or math.isnan(post_conf):
        return None

    return float(base_conf), float(post_conf)

# =========================
# Aggregation
# =========================
def summarize_means(group_by_bin: bool = False) -> pd.DataFrame:
    """
    Build a summary DataFrame of mean base and post-noise (same-choice) confidence
    for each K and model. Optionally split by confidence bin.
    Columns:
      ['K','Model','Bin'(opt),'N','BaseMean','PostMean']
    """
    rows = []
    for K in KS:
        for model in MODELS:
            if group_by_bin:
                for bin_key in BINS:
                    path = build_noise_path(K, model, bin_key)
                    data = read_json(path)
                    base_vals, post_vals = [], []
                    if data:
                        for item in data:
                            res = extract_base_and_post_conf(item)
                            if res is not None:
                                b, p = res
                                base_vals.append(b)
                                post_vals.append(p)
                    N = len(base_vals)
                    base_mean = float(np.mean(base_vals)) if N > 0 else np.nan
                    post_mean = float(np.mean(post_vals)) if N > 0 else np.nan
                    rows.append({
                        "K": K,
                        "Model": model,
                        "Bin": bin_key,
                        "N": N,
                        "BaseMean": base_mean,
                        "PostMean": post_mean
                    })
            else:
                # aggregate over all bins
                base_vals_all, post_vals_all = [], []
                for bin_key in BINS:
                    path = build_noise_path(K, model, bin_key)
                    data = read_json(path)
                    if not data:
                        continue
                    for item in data:
                        res = extract_base_and_post_conf(item)
                        if res is not None:
                            b, p = res
                            base_vals_all.append(b)
                            post_vals_all.append(p)
                N = len(base_vals_all)
                base_mean = float(np.mean(base_vals_all)) if N > 0 else np.nan
                post_mean = float(np.mean(post_vals_all)) if N > 0 else np.nan
                rows.append({
                    "K": K,
                    "Model": model,
                    "N": N,
                    "BaseMean": base_mean,
                    "PostMean": post_mean
                })
    df = pd.DataFrame(rows)
    return df

# =========================
# Plotting (matplotlib only)
# =========================
def plot_base_vs_post_by_model(summary_df: pd.DataFrame, K: int, save_dir: str, by_bin: bool = False) -> str:
    """
    Make a single figure (one chart) for a given K:
    x-axis = models, and two bars per model: BaseMean vs PostMean.
    Returns saved PNG path.
    """
    if by_bin:
        # Expect 'Bin' column present. Create one figure per bin for this K.
        out_paths = []
        for bin_key in BINS:
            sub = summary_df[(summary_df["K"] == K) & (summary_df["Bin"] == bin_key)].copy()
            if sub.empty:
                continue
            models = list(sub["Model"])
            base_vals = sub["BaseMean"].values
            post_vals = sub["PostMean"].values
            N_vals = sub["N"].values

            x = np.arange(len(models))
            width = 0.35

            fig, ax = plt.subplots(figsize=(10, 5))
            ax.bar(x - width/2, base_vals, width, label="Base (Mean)")
            ax.bar(x + width/2, post_vals, width, label=f"Post@{K} (Mean)")

            ax.set_title(f"Base vs Post@{K} (Mean) — Bin: {bin_key}")
            ax.set_xticks(x)
            ax.set_xticklabels(models, rotation=20, ha="right")
            ax.set_ylabel("Confidence (Mean)")
            ax.legend()

            # Annotate N on top of grouped bars (centered)
            for xi, n in zip(x, N_vals):
                ax.text(xi, max(base_vals.min(), post_vals.min())*0.02, f"N={int(n)}", rotation=90,
                        va="bottom", ha="center", fontsize=9)

            fig.tight_layout()
            out_path = os.path.join(save_dir, f"noise_base_post_mean_K{K}_bin_{bin_key}.png")
            fig.savefig(out_path, dpi=200, bbox_inches="tight")
            plt.close(fig)
            out_paths.append(out_path)
        return ", ".join(out_paths)
    else:
        sub = summary_df[summary_df["K"] == K].copy()
        if sub.empty:
            return ""

        models = list(sub["Model"])
        base_vals = sub["BaseMean"].values
        post_vals = sub["PostMean"].values
        N_vals = sub["N"].values

        x = np.arange(len(models))
        width = 0.35

        fig, ax = plt.subplots(figsize=(10, 5))
        ax.bar(x - width/2, base_vals, width, label="Base (Mean)")
        ax.bar(x + width/2, post_vals, width, label=f"Post@{K} (Mean)")

        ax.set_title(f"Base vs Post@{K} (Mean) — All Confidence Bins Combined")
        ax.set_xticks(x)
        ax.set_xticklabels(models, rotation=20, ha="right")
        ax.set_ylabel("Confidence (Mean)")
        ax.legend()

        # Annotate N per model below the bars
        for xi, n in zip(x, N_vals):
            ax.text(xi, 0.01, f"N={int(n)}", rotation=90, va="bottom", ha="center", fontsize=9)

        fig.tight_layout()
        out_path = os.path.join(save_dir, f"noise_base_post_mean_K{K}.png")
        fig.savefig(out_path, dpi=200, bbox_inches="tight")
        plt.close(fig)
        return out_path

# =========================
# Run
# =========================
summary = summarize_means(group_by_bin=GROUP_BY_BIN)
summary_csv = os.path.join(OUT_DIR, "noise_base_post_means.csv")
summary.to_csv(summary_csv, index=False)
print(f"Saved summary CSV to: {summary_csv}")
print(summary.head(10))

if GROUP_BY_BIN:
    for K in KS:
        paths = plot_base_vs_post_by_model(summary, K, save_dir=OUT_DIR, by_bin=True)
        print(f"K={K} plots saved to: {paths}")
else:
    for K in KS:
        p = plot_base_vs_post_by_model(summary, K, save_dir=OUT_DIR, by_bin=False)
        print(f"K={K} plot saved to: {p}")
# %%
```

```python
# --- Configure input path (edit this if needed) ---
INPUT_JSON = "/content/KGSweetSpot/Inference_results/data/inference_Mistral-7B-Instruct-v0.3.concept_net_No_Knowledge.dev.csqa.json"

import json, os, re
from pathlib import Path
import numpy as np

# ---------- helpers ----------
def load_items(path):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)
    if not isinstance(data, list):
        raise ValueError("Input JSON must be a list of records.")
    return data

def ensure_ids(items):
    """Force stable 1-based 'id' for every item (overwrite if present)."""
    for i, it in enumerate(items, start=1):
        it["id"] = str(i)

def norm(s):
    return re.sub(r"\s+", " ", str(s).strip().lower())

def get_pred_idx(it):
    """
    Determine predicted choice index.
    Priority:
      1) 'pred_idx' if valid
      2) 'pred' (text) aligned to 'cands'
      3) argmax over 'probs'
      4) first of 'ranked' if present
    Returns int or None.
    """
    cands = it.get("cands", [])
    if not isinstance(cands, list) or len(cands) == 0:
        return None

    # 1) explicit index
    if isinstance(it.get("pred_idx"), int):
        idx = it["pred_idx"]
        if 0 <= idx < len(cands):
            return idx

    # 2) 'pred' text
    if isinstance(it.get("pred"), str):
        try:
            return next(i for i, t in enumerate(cands) if norm(t) == norm(it["pred"]))
        except StopIteration:
            pass

    # 3) 'probs' argmax aligned to cands
    if isinstance(it.get("probs"), list) and len(it["probs"]) >= len(cands):
        try:
            arr = np.asarray(it["probs"][:len(cands)], dtype=float)
            return int(np.nanargmax(arr))
        except Exception:
            pass

    # 4) 'ranked' first item
    if isinstance(it.get("ranked"), list) and it["ranked"]:
        ranked = it["ranked"]
        # try exact text match first
        top = ranked[0].get("choice")
        if isinstance(top, str):
            try:
                return next(i for i, t in enumerate(cands) if norm(t) == norm(top))
            except StopIteration:
                pass
        # else fallback to highest prob entry inside 'ranked'
        try:
            best = max([(j, d.get("prob", float("nan"))) for j, d in enumerate(ranked)], key=lambda x: (x[1]))
            top_text = ranked[best[0]].get("choice", "")
            return next(i for i, t in enumerate(cands) if norm(t) == norm(top_text))
        except Exception:
            pass

    return None

def is_correct(it):
    """Compare predicted text vs gold answer text (normalized)."""
    cands = it.get("cands", [])
    ans   = it.get("answer", None)
    if not isinstance(cands, list) or not isinstance(ans, str):
        return None  # cannot decide

    pred_idx = get_pred_idx(it)
    if pred_idx is None or not (0 <= pred_idx < len(cands)):
        return None

    return norm(cands[pred_idx]) == norm(ans)

# ---------- run ----------
items = load_items(INPUT_JSON)
ensure_ids(items)

correct, incorrect, undecidable = [], [], []

for it in items:
    flag = is_correct(it)
    if flag is True:
        correct.append(it)
    elif flag is False:
        incorrect.append(it)
    else:
        undecidable.append(it)  # missing fields / cannot align

# ---------- write outputs ----------
base = Path(INPUT_JSON)
stem = base.stem
out_dir = base.parent

paths = {
    "correct":    out_dir / f"{stem}.correct.json",
    "incorrect":  out_dir / f"{stem}.incorrect.json",
    # "undecidable": out_dir / f"{stem}.undecidable.json",
}

with open(paths["correct"], "w", encoding="utf-8") as f:
    json.dump(correct, f, indent=2, ensure_ascii=False)
with open(paths["incorrect"], "w", encoding="utf-8") as f:
    json.dump(incorrect, f, indent=2, ensure_ascii=False)
# with open(paths["undecidable"], "w", encoding="utf-8") as f:
#     json.dump(undecidable, f, indent=2, ensure_ascii=False)

summary = {
    "input_file": str(base),
    "total_items": len(items),
    "correct": len(correct),
    "incorrect": len(incorrect),
    "outputs": {k: str(v) for k, v in paths.items()},
}
summary
```

```python
# BOTH models, GLOBAL noise only, in one figure per chart type (boxplots, heatmaps, line graphs)
# Rows: [Global noise (T5-small), Global noise (T5-large)]
# Cols: [low, mid, high]
#
# Outputs:
#   /content/results/mega/BOXPLOTS_global_small_vs_large.png
#   /content/results/mega/HEATMAPS_global_small_vs_large.png
#   /content/results/mega/TRENDS_global_small_vs_large.png
#
# Constraints: matplotlib only; one chart per figure; no explicit colors/styles.

import os, json, re
from pathlib import Path
from typing import Dict, List
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# File paths (as requested)
# -------------------------------
FILES_GLOBAL_SMALL = {  # T5-small (global noise)
    "low":  "/content/KGSweetSpot/Inference_results/data/noise_global/10_noisy_knowledge/T5_small/inference_flan-t5-small.noise_global.low.json",
    "mid":  "/content/KGSweetSpot/Inference_results/data/noise_global/10_noisy_knowledge/T5_small/inference_flan-t5-small.noise_global.medium.json",
    "high": "/content/KGSweetSpot/Inference_results/data/noise_global/T5_small/inference_flan-t5-small.noise_global.high.json",
}

FILES_GLOBAL_LARGE = {  # T5-large (global noise)
    "low":  "/content/KGSweetSpot/Inference_results/data/noise_global/10_noisy_knowledge/T5_large/inference_flan-t5-large.noise_global.low.json",
    "mid":  "/content/KGSweetSpot/Inference_results/data/noise_global/10_noisy_knowledge/T5_large/inference_flan-t5-large.noise_global.medium.json",
    "high": "/content/KGSweetSpot/Inference_results/data/noise_global/10_noisy_knowledge/T5_large/inference_flan-t5-large.noise_global.high.json",
}

ROW_TITLES = ["Global noise (T5-small)", "Global noise (T5-large)"]

OUTDIR = Path("/content/results/mega")
OUTDIR.mkdir(parents=True, exist_ok=True)

# -------------------------------
# Helpers
# -------------------------------
def norm(s: str) -> str:
    return re.sub(r"\s+", " ", str(s).strip().lower())

def load_rows(path: str):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

def probs_from_ranked(cands: List[str], ranked_list: List[Dict]) -> np.ndarray:
    mp = {norm(d.get("choice","")): float(d.get("prob", 0.0)) for d in ranked_list}
    vec = np.array([mp.get(norm(c), 0.0) for c in cands], dtype=float)
    s = vec.sum()
    return vec / s if s > 0 else vec

def order_from_ranked(cands: List[str], ranked_list: List[Dict]) -> List[int]:
    name2idx = {norm(c): i for i, c in enumerate(cands)}
    order = [name2idx[norm(d.get("choice",""))] for d in ranked_list if norm(d.get("choice","")) in name2idx]
    order += [i for i in range(len(cands)) if i not in set(order)]
    return order

def build_rows_for_case(files_case: Dict[str, str]) -> Dict[str, List[Dict]]:
    per_bucket = {"low": [], "mid": [], "high": []}
    for bucket, path in files_case.items():
        if not os.path.exists(path):
            continue
        rows = load_rows(path)
        for it in rows:
            cands = it.get("cands", [])
            if not cands:
                continue
            ranked_base = it.get("ranked", [])
            ranked_with = it.get("ranked_with_ks", [])
            if ranked_base and ranked_with:
                base_probs = probs_from_ranked(cands, ranked_base)
                with_probs = probs_from_ranked(cands, ranked_with)
                base_order = order_from_ranked(cands, ranked_base)
                with_order = order_from_ranked(cands, ranked_with)
            else:
                base_probs = np.array(it.get("base_probs", []), dtype=float)
                with_probs = np.array(it.get("probs", []), dtype=float)
                if base_probs.size != len(cands) or with_probs.size != len(cands):
                    continue
                if base_probs.sum() > 0: base_probs /= base_probs.sum()
                if with_probs.sum() > 0: with_probs /= with_probs.sum()
                base_order = list(np.argsort(-base_probs))
                with_order = list(np.argsort(-with_probs))

            base_pos = {i: base_order.index(i) for i in range(len(cands))}
            with_pos = {i: with_order.index(i) for i in range(len(cands))}
            for i in range(len(cands)):
                per_bucket[bucket].append({
                    "base_prob": float(base_probs[i]),
                    "with_prob": float(with_probs[i]),
                    "delta_prob": float(with_probs[i] - base_probs[i]),
                    "base_rank": base_pos[i],
                    "with_rank": with_pos[i],
                })
    return per_bucket

def transition_matrix(rows, K=5):
    M = np.zeros((K, K), dtype=int)
    for r in rows:
        rb, rw = int(r["base_rank"]), int(r["with_rank"])
        if 0 <= rb < K and 0 <= rw < K:
            M[rb, rw] += 1
    return M

def median_series(rows, key: str):
    out = []
    for r in range(5):
        vals = [item[key] for item in rows if int(item["base_rank"]) == r]
        out.append(float(np.median(vals)) if len(vals) else np.nan)
    return np.array(out, dtype=float)

# -------------------------------
# Load both models (global noise)
# -------------------------------
rows_small = build_rows_for_case(FILES_GLOBAL_SMALL)
rows_large = build_rows_for_case(FILES_GLOBAL_LARGE)
buckets = ["low", "mid", "high"]

# -------------------------------
# Figure 3: Median P vs base rank (2×3)
# -------------------------------
fig3, axs3 = plt.subplots(2, 3, figsize=(16, 9), sharey=True, sharex=True)
for r, case_rows in enumerate([rows_small, rows_large]):
    for c, bucket in enumerate(buckets):
        ax = axs3[r, c]
        rows = case_rows.get(bucket, [])
        x = np.arange(1, 6)
        if rows:
            base_med = median_series(rows, "base_prob")
            with_med = median_series(rows, "with_prob")
            ax.plot(x, base_med, label="base")
            ax.plot(x, with_med, label="with-noise")
        else:
            ax.text(0.5, 0.5, "No data", ha="center", va="center", transform=ax.transAxes)
        ax.set_xticks(x); ax.set_xticklabels([str(i) for i in x])
        if r == 1:
            ax.set_xlabel("Base rank (1 = top)")
        if c == 0:
            ax.set_ylabel("Median P(choice)")
        ax.set_title(f"{ROW_TITLES[r]} — {bucket}")
handles, labels = axs3[0,0].get_legend_handles_labels()
if handles:
    fig3.legend(handles, labels)
fig3.suptitle("Median P vs base rank — Global noise (T5-small vs T5-large)", y=0.98)
fig3.tight_layout()
fig3.savefig(OUTDIR / "TRENDS_global_small_vs_large.png", dpi=150)
plt.close(fig3)

print("Saved:")
print(" -", OUTDIR / "TRENDS_global_small_vs_large.png")
```
